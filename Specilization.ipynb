```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go

data = pd.read_csv("/content/Updated_FitBit_data_with_modified_age_group.csv")
print(data)
```

```python
print(data.isnull().sum())
```

```python
print(data.info())
```

```python
data["ActivityDate"] = pd.to_datetime(data["ActivityDate"],
                                format="%m/%d/%Y")

print(data.info())
```

```python
data["TotalMinutes"] = data["VeryActiveMinutes"] + data["FairlyActiveMinutes"] + data["LightlyActiveMinutes"] + data["SedentaryMinutes"]
print(data["TotalMinutes"].sample(5))
```

```python

def assign_health_metrics(row):
    age = np.random.choice(['18-24', '60+', '35-59'])
    if age == '18-24':
        heart_rate = np.random.randint(60, 121)
        ecg = np.random.randint(60, 101)
        stress_level = np.random.randint(1, 6)
    elif age == '60+':
        heart_rate = np.random.randint(60, 111)
        ecg = np.random.randint(60, 121)
        stress_level = np.random.randint(3, 8)
    else:
        heart_rate = np.random.randint(70, 131)
        ecg = np.random.randint(80, 130)
        stress_level = np.random.randint(6, 11)

    return pd.Series([age, heart_rate, ecg, stress_level])

data[['Age Group', 'Heart Rate', 'ECG', 'Stress Level']] = data.apply(assign_health_metrics, axis=1)

data.head()
```

```python
fig = px.scatter(
    data,
    x='Calories',
    y='TotalSteps',
    size='Heart Rate',  # Point size based on heart rate
    color='Age Group',  # Color points by age group
    hover_data=['ECG', 'Stress Level'],  # Show ECG
    title="Relationship between Calories & Total Steps with Health Metrics"
)
fig.show()
```
